```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
from math import sqrt
```

```python
#load the dataset
data= pd.read_excel('fruit_types.xlsx')
data.head()
```

```python
data.info()
```

```python
#label encoder
encoder= LabelEncoder()

#encode all the variables in the dataset
fruit_name_encoded = encoder.fit_transform(data.iloc[:,0])
fruit_subtype_encoded = encoder.fit_transform(data.iloc[:,1])
mass_encoded = encoder.fit_transform(data.iloc[:,2])
width_encoded = encoder.fit_transform(data.iloc[:,3])
height_encoded = encoder.fit_transform(data.iloc[:,4])


```

```python
#combine encoded data
data = list(zip(fruit_name_encoded, fruit_subtype_encoded, mass_encoded,width_encoded,height_encoded))

#put data into a data frame
data = pd.DataFrame(data,columns=['fruit_name','fruit_subtype','mass','width','height'])
data.head()
```

```python
#test train split
data_to_use = data.iloc[:,1:5]
data_to_target =data.iloc[:,0]
```

```python
# split , test our data
X_train,X_test,y_train,y_test =train_test_split(data_to_use,data_to_target, test_size=0.3, random_state=32)
```

```python
#build kNN model
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train,y_train)
```

```python
#make prediction
y_pred_knn = knn_model.predict(X_test)
```

```python
print('Accuracy:',metrics.accuracy_score(y_test,y_pred_knn))
```

```python
print('RMSE:',metrics.mean_squared_error(y_test,y_pred_knn))
```

```python
#decision tree
#gini
#max depth
#entropy
decision_gini = DecisionTreeClassifier()
decision_entropy = DecisionTreeClassifier(criterion='entropy')
decision_depth =DecisionTreeClassifier(max_depth=5)
```

```python
decision_gini.fit(X_train,y_train)
```

```python

decision_entropy.fit(X_train,y_train)
```

```python
decision_depth.fit(X_train,y_train)
```

```python
y_pred_gini =decision_gini.predict(X_test)
y_pred_entropy =decision_entropy.predict(X_test)
y_pred_depth =decision_depth.predict(X_test)
```

```python
print('Accuracy(gini):',metrics.accuracy_score(y_test,y_pred_gini))
print('Accuracy(depth):',metrics.accuracy_score(y_test,y_pred_depth))
print('Accuracy(entropy):',metrics.accuracy_score(y_test,y_pred_entropy))
```

```python
## create an object
clflinear= svm.SVC(kernel='linear')
clfsigmoid=svm.SVC(kernel='sigmoid')
clfrbf = svm.SVC(kernel='rbf')
```

```python
clflinear.fit(X_train,y_train)
```

```python
clfsigmoid.fit(X_train,y_train)
```

```python
clfrbf.fit(X_train,y_train)
```

```python
y_pred_lin = clflinear.predict(X_test)
y_pred_sig = clfsigmoid.predict(X_test)
y_pred_rbf = clfrbf.predict(X_test)
```

```python
print('Accuracy(lin):',metrics.accuracy_score(y_test,y_pred_lin))
print('Accuracy(sig):',metrics.accuracy_score(y_test,y_pred_sig))
print('Accuracy(rbf):',metrics.accuracy_score(y_test,y_pred_rbf))
```
