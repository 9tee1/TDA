```python
#Import all the required libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import plotly.express as px
```

```python
#Load dataset using Pandas function pd.read_csv
df = pd.read_csv('all_month.csv')

#Visualise a sample of the dataset
df.head(10)
```

```python
# From the above output we can see the way that the data is structured. Each line represents an occurance of an earthquake
```

```python
##Get information related to our dataset
df.info()
```

```python
#From the above output we can see that there are a total of 8886 occurances of earthquakes within our dataset.
#We can see there are 22 columns, each representing different attributes related to the earthquake.  
#We can also identify that some of the values are missing. This is shown by the number of 'non-null' values.
```

```python
#The below function allows us to determine the total number of missing values in each column
df.isnull().sum()
```

```python
# We now need to replace the empty values with 'NaN'(Not a Number)
# We can do this by using the below replace function
df=df.replace({'':np.nan})
df.head()
```

```python
#We can now use the '.fillna' function to replace all values that are showing as NaN in our dataframe. 
# The missing values will be replaced with the median from each column as this is common practice and reduces bias and error.
#This also ensures compatability with algorithms that are unable to handle missing values.
df['nst'] = df['nst'].fillna(df['nst'].median())
df['gap'] = df['gap'].fillna(df['gap'].median())
df['dmin'] = df['dmin'].fillna(df['dmin'].median())
df['horizontalError'] = df['horizontalError'].fillna(df['horizontalError'].median())
df['magError'] = df['magError'].fillna(df['magError'].median())
df['magNst'] = df['magNst'].fillna(df['magNst'].median())
```

```python
#We can now use the same function used earlier to verify there are no missing values
df.isnull().sum()
```

```python
#Another way to visualise the data is to use histograms. This shows the frequency distribution of a continuous variable.
#This can be used to better understand the shape, outliers and also patterns within our data.
df.hist(bins=25, figsize=(15,10), color= 'Green',)
plt.show()
```

```python
#We are only able to use this plot for any values that are numerical.
```

```python
#Derive the date using the time ISO 8601 format into a newdate column
df['date'] = pd.DatetimeIndex(df['time']).date
#Derive the month from the new date column
df['month'] = pd.DatetimeIndex(df['date']).month
#Derive the year from the new date column
df['year'] = pd.DatetimeIndex(df['time']).year
```

```python
#Show the desired result with the new columns appended to the right of the table
df.head()
```

```python
# Find the earliest and latest date in our dataset
earliest_date = df['date'].min()
latest_date = df['date'].max()

print(earliest_date)
print(latest_date)
```

```python
#Pie Chart showing the percentage of earthquakes for each year
plt.figure(figsize=(10, 6))
#Count the number of occurrences of each unique valu in the year column and plot the chart
df['year'].value_counts().plot(kind='pie', autopct='%1.2f%%')
#Set the title
plt.title('Percentage of Earthquakes per Year')
#Show the graph
plt.show()
```

```python
#The above chart shows that our dataset has almost the same number of earthquakes in the 2023(Dec) and 2024(Jan) 
```

```python
#A bar graph showing the type of earthquakes in our dataset
plt.figure(figsize=(10, 6))
#Count the number of occurrences of each unique value in the type column and plot the chart
df['type'].value_counts().plot(kind='barh')
#Set the title
plt.title('Types of Earthquakes')
plt.xlabel('Number of occurences')
#Show the graph
plt.show()
```

```python
#Create a line graph showing the average depths by date
# Group by the date and calculate the mean depth
x = df.groupby('date')['depth'].mean()

# Create a figure
fig = plt.figure(figsize=(10, 5))

# Create a line plot showing the average depth of earthquakes by date
sns.lineplot(data=x)

# Set the title and labels
plt.title('Average Depth of Earthquakes by Date')
plt.xlabel('Date')
plt.ylabel('Depth (km)')

# Display the plot
plt.show()

# Adjust the layout
fig.tight_layout()
```

```python
#From the above plot we can establish that as time goes on the depth of the earthquakes has become reduced(closer to surface level).
#This could mean that there has been an increase in shallow seismic activity and/or changes in tectonic activity.
```

```python
# Calculate the 30 most common places where an earthquake has occurred
top_places = df['place'].value_counts().head(30)

# Create a horizontal bar plot
plt.figure(figsize=(10, 10))
sns.barplot(x=top_places.values, y=top_places.index, orient='h',color='purple')
plt.title('A bar chart showing 30 of the most common places where an earthquake has occurred')
plt.xlabel('Number of occurrences')
plt.show()
```

```python
#From the above chart we can establish that the most common place an earthquake has occured in our data is 8 km NW of The Geysers,CA.
#from this we can also establish the most active regions within our dataset.
```

```python
# Create a scatter plot with a regression line to show the correlation between Mag and Depth
plt.figure(figsize=(10, 5))
sns.regplot(data=df, 
            x='mag', 
            y='depth',
            ci=97, #ci=confidence interval
            marker='+',
            color='grey',
            scatter_kws={'alpha':0.5},
            line_kws={"color": 'red'}) 
plt.title('Relationship between Depth and Magnitude of Earthquakes')
plt.ylabel('Depth (km)')
plt.xlabel('Magnitude')
plt.show()
```

```python
#The scatter plot allows us to observe any potential correlation or trend between the depth and magnitude of earthquakes. 
#From the above plot we can see that there isn't much of a correlation between the magnitude of earthquakes that happen at a depth between 0-250 km.
#We can also say that the earthquakes at depths of 250-300 are mostly high magnitude earthquakes, however, this assumption has a lower 
#certainty taking into consideration the magnitudes vs depth between 0 - 250 km.

#The regression line has a higher confidence interval at depths of 0-250 km and a lower certainty at depths between 250 km+.
```

```python
#create a boxplot showing any outliers in our data
# Drop the 'year' and 'month' columns
df_new = df.drop(['year', 'month'], axis=1)

# Select only the numerical columns
df_new = df_new.select_dtypes(include=[np.number])

# Create the boxplots
plt.figure(figsize=(14, 10))  # Increase figure size
plt.boxplot(df_new.values, vert=False, patch_artist=True, sym='b+')  # Change boxplot color
plt.title('Box Plot Showing Outliers in Earthquake Data', loc="center")
plt.yticks(range(1, len(df_new.columns) + 1), df_new.columns)  # Add y-axis labels
plt.grid(True)  # Add gridlines
plt.gca().set_facecolor('none')
plt.show()
```

```python
#The above boxplot shows the distribution of data based on a five-number summary 
#(“minimum”, first quartile (Q1), median, third quartile (Q3), and “maximum”).
```

```python
# Create the Geo scatter plot
fig2 = px.scatter_geo(df, lat='latitude',
                     lon='longitude',
                     color="mag",
                     fitbounds='locations', #This ensures that the geographical area in the plot will be zoomed and centred to show all visible plots.
                     scope='world') #This argument sets the scope of the map. Currently set to the whole world, however, can be changed. 

# Update the layout size
fig2.update_layout(
    autosize=False,
    width=700,  # Set the width
    height=600)  # Set the height

# Display the plot
fig2.show()
```

```python
#The above is an interactive Geo scatter plot of our data. Each point’s location is determined by the ‘latitude’ and ‘longitude’ columns, 
#each point’s color determined by the ‘mag’ column, and the map’s scope set to the entire world. This can also be changed to show specific 
#regions of interest.
#From the above scatter plot we can better visualise the earthquake location and which areas are more active than others.
```
